6. 발전전력량 추이

```python
data = energy_generations_by_year_month
data.set_index("월별(1)", inplace=True)
data.index.name = "월"
data
```

```python
data_list = []

for year in range(2015, 2023):
    year_columns = [column for column in data.columns if str(year) in column]
    data_list.append(data[year_columns])
```

```python
for data in data_list:
    new_columns = [i + " " + j + " " + k for i,j,k in zip(data.iloc[0], data.iloc[1], data.iloc[2])]
    data.columns = new_columns
    data.drop(data.index[:3], inplace=True)
data_list[0]
```

```python
for data in data_list:
    for column in data.columns:
        if "소계 소계" in column:
            data.columns = [i.replace("소계 소계", "총계") for i in data.columns]
            
data_list[0]
```
